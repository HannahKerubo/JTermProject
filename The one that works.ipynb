```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'CTYD Kenya.csv'
data = pd.read_csv(file_path)
```

```python
#| label: GRPtable
data[ ['region', 'year', 'grp_lcu'] ].head()
```

```python
pivot_data = data.pivot_table(values='grp_lcu', index='region', columns='year', aggfunc='sum')
```

```python
#| label: GRPchart
pivot_data.plot(kind='bar', stacked=True, color=['blue','orange'], figsize=(15, 8))
plt.title('Gross Regional Product (grp_lcu) per Region in Kenya for 2016 and 2017')
plt.xlabel('Region')
plt.ylabel('Gross Regional Product (grp_lcu)')
plt.xticks(rotation=90)
plt.legend(title='Year')
plt.savefig('GRPGRAPH.png')
```
